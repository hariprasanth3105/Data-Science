```python
import numpy as np
y1=blr.loc[(blr['day']=='Saturday'),['day','congestion level']]
d1=y1['congestion level'].sum()

y2=blr.loc[(blr['day']=='Sunday'),['day','congestion level']]
d2=y2['congestion level'].sum()

y3=blr.loc[(blr['day']=='Monday'),['day','congestion level']]
d3=y3['congestion level'].sum()

y4=blr.loc[(blr['day']=='Tuesday'),['day','congestion level']]
d4=y4['congestion level'].sum()

y5=blr.loc[(blr['day']=='Wednesday'),['day','congestion level']]
d5=y5['congestion level'].sum()

y6=blr.loc[(blr['day']=='Thursday'),['day','congestion level']]
d6=y6['congestion level'].sum()

y7=blr.loc[(blr['day']=='Friday'),['day','congestion level']]
d7=y7['congestion level'].sum()

y_col=np.array([d1,d2,d3,d4,d5,d6,d7])
print(y_col)
```

#### There is no significant difference. However, the slight downward trend observed on Wednesday, Thursday, and Friday suggests that there might be a minor impact of the day of the week on congestion.

### Analyse how season affect the traffic congestion.|

```python
#add a new column 'year'
blr['date']=pd.to_datetime(blr['date'],format=('%Y-%m-%d'))
blr['year']=blr['date'].dt.year
#add year column to the 3rd column
ycol=blr.columns.tolist()
ycol
ly=ycol.pop()
ycol.insert(2,ly)
blr=blr[ycol]
```

```python
blr22_23 = blr[blr['year'] <= 2023].copy()
 #The .copy() ensures that this new DataFrame is an independent copy, preventing any unintentional changes to the original data.
print(blr22_23.info())
```

```python

#blr22_23 = blr[blr['year'] <= 2023].copy()
#The .copy() ensures that this new DataFrame is an independent copy, preventing any unintentional changes to the original data.

import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

#we'll categrosize the month
month_od22_23=['January', 'February', 'March', 'April', 'May', 'June', 'July','August', 'September', 'October', 'November', 'December']
blr22_23['month'] = pd.Categorical(blr22_23['month'], categories=month_od22_23, ordered=True)

month_grp=blr22_23.groupby('month',observed=False)['congestion level'].sum()

#graph
m_color=['blue','blue','orange','orange','orange','green','green','green','green','lightgreen','lightgreen']

plt.bar(month_grp.index,month_grp.values,color=m_color)
plt.plot(month_grp.index,month_grp.values,linestyle="--",color='red',marker='o',alpha=0.7)
plt.xticks(rotation=70)
plt.xlabel('Months')
plt.ylabel('Congestion level')
plt.title('Monthly Traffic Congestion Trends (2022-2023) : A Seasonal Analysis"')
legend_elements = [Patch(facecolor='blue', label='Winter'),
                   Patch(facecolor='orange', label='Summer'),
                   Patch(facecolor='green', label='Monsoon'),
                   Patch(facecolor='lightgreen', label='Post Monsoon')]

# Add the legend to the plot
plt.legend(handles=legend_elements, bbox_to_anchor=(1.05, 1), loc='upper left')
#month_grp = month_grp.sort_index()
plt.show()
```

**Key Insights :**
>***Seasonal Variation:*** The graph suggests that congestion is lowest during the Winter months and highest during the Summer and Monsoon seasons. This could be due to weather conditions, holidays, or other seasonal factors that affect traffic patterns.

>***Peak Congestion:*** April, which falls in the Summer season, is the peak congestion month, possibly due to increased travel during the season.

>***Consistent Congestion:*** The Monsoon season shows high but steady congestion, which could be influenced by weather conditions such as rain, leading to slower traffic movement.

>***Environmental Impact:*** Koramangala consistently has a higher environmental impact, likely due to its higher traffic volume and congestion levels.

>***Congestion and Travel Time:*** Koramangala also has higher congestion and longer travel times, impacting overall traffic efficiency.

***Conclusion:***
>This heatmap reveals that Koramangala tends to have higher traffic volumes, parking usage, and environmental impact than Indiranagar, which aligns with its busier and more commercially dense character. Indiranagar, while also urban, seems to be more pedestrian-friendly with lower traffic volumes and a higher count of cyclists and pedestrians.
```python
import matplotlib.pyplot as plt
import pandas as pd

areas=['Indiranagar', 'M.G. Road', 'Koramangala', 'Jayanagar','Hebbal', 'Whitefield', 'Yeshwanthpur', 'Electronic City']
blr['area name']=pd.Categorical(blr['area name'],categories=areas,ordered=True)

area_grp2 = blr.groupby('area name', observed=False)[['public transport usage', 'environmental impact']].sum()

ptu=area_grp2.values[:,0]
ei=area_grp2.values[:,1]

# Plotting the data
plt.figure(figsize=(14, 7))

# Bar chart for Public Transport Usage
plt.bar(area_grp2.index,ptu, color='skyblue', label='Public Transport Usage')

# Bar chart for Road Capacity Utilization (stacked on top)
plt.bar(area_grp2.index,ei, bottom=ptu, color='lightcoral', label='Environmental Impact')

# Adding titles and labels
plt.title('Public Transport Usage and Environmental Impact by Area', fontsize=16)
plt.xlabel('Area Name', fontsize=12)
plt.ylabel('Usage / Impact', fontsize=12)
plt.xticks(rotation=45)  # Rotating x labels for better readability

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```

**Key Insights:**
>***Indiranagar and M.G.*** Road have the highest overall values, indicating both high public transport usage and a significant environmental impact.

>***Electronic City*** has the lowest values, suggesting lower public transport usage and environmental impact compared to other areas.
